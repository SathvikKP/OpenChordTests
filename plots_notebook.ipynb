```python
import numpy as np
import matplotlib.pyplot as plt


peers = [4, 8, 16, 32, 64]
lookup_time_p1 = [86, 114, 209, 253, 0]
same_peer_again = [86, 111, 131, 211, 0]
random_peer = [85, 106, 158, 238, 0]
random_again = [85, 101, 133, 204, 0]

plt.figure(figsize=(10, 6))
bar_width = 0.2
x = np.arange(len(peers))

plt.bar(x - 1.5 * bar_width, lookup_time_p1, width=bar_width, label="Lookup Time P1", color="b")
plt.bar(x - 0.5 * bar_width, same_peer_again, width=bar_width, label="Same Peer Again", color="g")
plt.bar(x + 0.5 * bar_width, random_peer, width=bar_width, label="Random Peer", color="r")
plt.bar(x + 1.5 * bar_width, random_again, width=bar_width, label="Random Again", color="orange")

plt.xticks(x, peers)
plt.xlabel("Number of Peers (n)")
plt.ylabel("Lookup Time (ms)")
plt.title("Effect of Number of Peers on Lookup Times")
plt.legend()
plt.savefig("table1_lookup_times.png")
plt.show()


successors = [1, 2, 4, 8]
lookup_time_p1_succ = [298, 286, 276, 253]
same_peer_again_succ = [232, 234, 234, 211]
random_peer_succ = [271, 248, 243, 238]
random_again_succ = [232, 211, 209, 204]

plt.figure(figsize=(10, 6))
x = np.arange(len(successors))

plt.bar(x - 1.5 * bar_width, lookup_time_p1_succ, width=bar_width, label="Lookup Time P1", color="b")
plt.bar(x - 0.5 * bar_width, same_peer_again_succ, width=bar_width, label="Same Peer Again", color="g")
plt.bar(x + 0.5 * bar_width, random_peer_succ, width=bar_width, label="Random Peer", color="r")
plt.bar(x + 1.5 * bar_width, random_again_succ, width=bar_width, label="Random Again", color="orange")

plt.xticks(x, successors)
plt.xlabel("Number of Successors")
plt.ylabel("Lookup Time (ms)")
plt.title("Impact of Number of Successors on Lookup Times (32 peers)")
plt.legend()
plt.savefig("table2_successors.png")
plt.show()


extra_insertions = [0, 32, 96, 224, 480]
total_insertions = [32, 64, 128, 256, 512]
lookup_time = [240, 230, 212, 205, 195]
additional_insertion_time = [0, 261, 192, 201, 187]

plt.figure(figsize=(10, 6))
x = np.arange(len(extra_insertions))

plt.bar(x - bar_width / 2, lookup_time, width=bar_width, label="Lookup Time (ms)", color="b")
plt.bar(x + bar_width / 2, additional_insertion_time, width=bar_width, label="Additional Insertion Time (ms)", color="r")

plt.xticks(x, total_insertions)
plt.xlabel("Total Insertions")
plt.ylabel("Time (ms)")
plt.title("Effect of Extra Insertions on System Performance")
plt.legend()
plt.savefig("table3_insertions.png")
plt.show()
```
